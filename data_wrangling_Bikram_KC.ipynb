````python
### Ex-7-Task-2

df_sales['total_profit'] = None
# Calculate the total profit using apply function in pandas and save in total_profit column
# Formulae: total_profit = (mean product price - mean retail price) * total units sold
### BEGIN SOLUTION
# Step 1: Calculate the required aggregate values
mean_price = df_product['price'].mean()
mean_retail_price = df_product['retail_price'].mean()
total_units_sold = df_product['units_sold'].sum()
# Step 2: Use apply (though apply is not needed, we'll use it as requested)
df_sales['total_profit'] = df_sales.apply(
    lambda row: (mean_price - mean_retail_price) * total_units_sold,
    axis=1
)
### END SOLUTION
````

````python
### Ex-8-Task-2
group_df = None
## Calculate the Coefficient of Variation of each sales category
### BEGIN SOLUTION
group_df = df_product.groupby('sales_category')['price'].agg(['mean', 'std'])
group_df['coefficient_of_variation'] = (group_df['std'] / group_df['mean']) * 100
### END SOLUTION
````

````python
### Ex-8-Task-3
df_product['price_zscore'] = None
## Calculate the Z_score of price for each sales category
### BEGIN SOLUTION
# Define a function to compute z-score within each group
def calculate_zscore(group):
    mean_price = group['price'].mean()
    std_price = group['price'].std()
    group['price_zscore'] = (group['price'] - mean_price) / std_price
    return group
# Apply the function to each group
df_product = df_product.groupby('sales_category').apply(calculate_zscore)
### END SOLUTION
````

## Product Rating and Performance Comparision

Looking at the rating columns I notice two things:

- Some products have a rating count that's too low to be considered, let's only look at products with more than 150 ratings to try to have a fair result
- The rating of the products is a decimal number: let's create another column that categorizes each product into it's rating range.
    - columns are
        1. < 1 if rating < 1
        2. 1-2 if rating >= 1 and rating < 2
        3. 2-3 if rating >= 2 and rating < 3
        4. 3-4 if rating >= 3 and rating < 4
        5. 4-5 if rating >= 4 and rating < 5
- Create pivot table with performance in index, raiting_range in column and no. of product as values



````python
### Ex-10-Task-5
df_ratings = pd.DataFrame()  ## get records with rating_count > 100
df_ratings['rating_range'] = None
pivot_table = None
### BEGIN SOLUTION
# Filter products with rating_count > 150
df_ratings = df[df['rating_count'] > 150].copy()
# Define function to assign rating ranges
def rating_range(r):
    if r < 1:
        return '<1'
    elif 1 <= r < 2:
        return '1-2'
    elif 2 <= r < 3:
        return '2-3'
    elif 3 <= r < 4:
        return '3-4'
    elif 4 <= r < 5:
        return '4-5'
    else:
        return None
# Apply rating range function
df_ratings['rating_range'] = df_ratings['rating'].apply(rating_range)
# Create pivot table with performance as index, rating_range as columns and count of products as values
pivot_table = pd.pivot_table(df_ratings, index='performance', columns='rating_range',
                             values='title', aggfunc='count', fill_value=0)
pivot_table
### END SOLUTION
````

This table shows that if the product have average rating of 3-4. THen, it is most likely to be successful product.
